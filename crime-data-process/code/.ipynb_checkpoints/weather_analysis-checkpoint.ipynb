```python
crime_2011 = sc.textFile('file:/Users/zhuorulin/Documents/DataScience/big-data-project/crime-data-process/results/data-exploration/weather/all_2011.out')
```

```python
date_format = '%Y/%m/%d'
dt.datetime.strptime('2011/01/01',date_format)
```

```python
date_format_crime = '%m/%d/%Y'
crime_2011_processed = crime_2011.map(lambda x:x.split('\t'))\
.map(lambda x:[dt.datetime.strptime(x[0],date_format_crime),int(x[1])])
crime_2011_processed_np = np.array(crime_2011_processed.collect())
```

```python
weather_process = weather_txt.map(lambda x:x.split('\t')).map(lambda x: [dt.datetime.strptime(x[0],date_format),x[1].split(',')])\
.map(lambda x: [x[0],float(x[1][0]),float(re.sub(string=x[1][1],pattern='[A-Z]',repl=''))])
weather_process_np = np.array(weather_process.collect())
```

```python
weather_process.take(10)
```

```python
temp_crime = weather_process.join(crime_2011_processed).collect()
```

```python
temp_crime = np.array([[line[1][0],line[1][1]] for line in temp_crime])
temp = temp_crime[:,0]
crime = temp_crime[:,1]
fit = np.polyfit(temp,crime,deg=1)
```

```python
plt.scatter(temp,crime)
plt.plot(temp, fit[0] * temp + fit[1], color='red')
plt.xlabel('Temperature(F)')
plt.ylabel('Day crime count')
plt.title('Temperature-Crime Relation (corr=0.37)')
plt.savefig('temperature_crime.png')
plt.ylim(800,1800)
plt.savefig('temp_crime_corr.png')
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
figure = ax1.get_figure()
figure.set_size_inches((10,5))
ax1.plot(weather_process_np[:,0],weather_process_np[:,1],label='temperature')
ax2.plot(crime_2011_processed_np[:,0],crime_2011_processed_np[:,1],label='crime',color='r')
ax1.set_ylabel('Temeprature(F)')
ax2.set_ylabel('Day Crime')
# ax2.set_ylim(0,5000)
# ax1.set_ylim(0,200)
plt.savefig('crime_temperature_trend.png')
```
