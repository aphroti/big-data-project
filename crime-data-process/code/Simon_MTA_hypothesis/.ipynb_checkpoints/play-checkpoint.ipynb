```python
import geojson
import geopy
import numpy as np
from geopy.distance import vincenty
import matplotlib.pyplot as plt
```

```python
f= open('./Subway Entrances.geojson','r')
mta_entrances_json = geojson.load(f)

mta_entrances = mta_entrances_json.features

mta_entrances_cdns = [point.geometry.coordinates for point in mta_entrances]

mta_entrances_cdns = np.array(mta_entrances_cdns)
```

```python
testx = crime_coordinates[0]
energy = [1/(vincenty(testx,station).km)**2 for station in mta_entrances_cdns]
```

```python
def cal_potential(x,stations_coordinates = mta_entrances_cdns):
    potential_list = [1/(vincenty(x,station).km)**2 for station in stations_coordinates]
    return sum(potential_list)
```

```python
latitude = crime_coordinates[:,0]
longitude = crime_coordinates[:,1]
heatmap, xedges, yedges = np.histogram2d(longitude,latitude, bins=100)
```

```python
np.sum(heatmap>0)
```

```python
energy = []
count_list = []
for i in range(100):
    for j in range(100):
        if j%10==0:
            print('i=%s,j=%s'%(i,j))
        count = heatmap[i,j]
        if count==0:
            continue
        longitude = (xedges[i]+xedges[i+1])/2
        latitude = (yedges[j]+yedges[j+1])/2
        energy.append(cal_potential([longitude,latitude]))
        count_list.append(count)
```

```python
plt.scatter(energy,count_list)
```

```python
energy = np.array(energy)
count_list = np.array(count_list)
np.corrcoef(energy,count_list)
```
