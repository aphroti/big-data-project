```python
def cal_potential(x,stations_coordinates = mta_entrances_cdns):
    potential_list = [1/(vincenty(x,station).km)**2 for station in stations_coordinates]
    return sum(potential_list)
```

```python
latitude = crime_coordinates[:,0]
longitude = crime_coordinates[:,1]
heatmap, xedges, yedges = np.histogram2d(longitude,latitude, bins=100)
```

```python
energy = []
count_list = []
idx=0
bar = progressbar.ProgressBar(max_value=10000)
energy_map= np.zeros([100,100])
for i in range(100):
    for j in range(100):
        idx+=1
        bar.update(idx)
        count = heatmap[i,j]
        if count==0:
            continue
        latitude = (xedges[i]+xedges[i+1])/2
        longitude = (yedges[j]+yedges[j+1])/2
        energy.append(cal_potential([longitude,latitude]))
        count_list.append(count)
        energy_map[i,j] = energy[-1]
```

```python
energy_map
```

```python
convolve_heat = convolve(energy_map, Gaussian2DKernel(stddev=2))
plt.figure(figsize=(10,10))
cax = plt.imshow(convolve(energy_map.T, Gaussian2DKernel(stddev=1))**0.3, interpolation='none')
plt.savefig('mta_energy.png')
plt.show()
```

```python
plt.figure(figsize=(15,5))
df_2plot = pd.DataFrame()
df_2plot['energy'] = energy
df_2plot['gridcount'] = count_list
df_2plot.plot.scatter('energy','gridcount',ylim=(0,1500),xlim=(0,500))
```
