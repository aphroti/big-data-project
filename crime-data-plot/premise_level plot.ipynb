```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
f = open('../crime-data-process/results/premise_level.out','rb')
lines= f.read().strip().decode('utf-8').split('\n')
f.close()
pairs = [line.split('\t') for line in lines]
areas = [pair[0].split(', ')[0] for pair in pairs]
types = [pair[0].split(', ')[1] for pair in pairs]
values = []
for pair in pairs:
    try:
        values.append(int(pair[1]))
    except ValueError:
        values.append(0)
df = pd.DataFrame([[area,types[i], values[i]] for i,area in enumerate(areas)],\
                  columns=['area','type','count'])
#df = df.sort_values(by=['year','month'])
print(df)
df = df.pivot(index='area',columns='type',values='count')
print(df)
```

```python
df.plot.bar(stacked=True, figsize = (25, 15))
plt.xlabel
plt.savefig('premise_level')
plt.show()
```
