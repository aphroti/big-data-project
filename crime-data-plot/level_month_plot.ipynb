```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
f = open('../crime-data-process/results/level_month/level_month.out','rb')
lines= f.read().strip().decode('utf-8').split('\n')
f.close()
pairs = [line.split('\t') for line in lines]
keys = [pair[0].split(' ')[0] for pair in pairs]
months = [pair[0].split(' ')[1].zfill(2) for pair in pairs]
years = [int(pair[0].split(' ')[2]) for pair in pairs]
values = [int(pair[1]) for pair in pairs]
df = pd.DataFrame([[key,months[i],years[i], values[i]] for i,key in enumerate(keys)],\
                  columns=['type','month','year','count'])
df = df.sort_values(by=['year','month'])
#print(df)
df['date'] = df['year'].map(str) + ' ' + df['month'].map(str)
df = df.pivot(index='date',columns='type',values='count')
#print(df)
```

```python
#plt.figure(figsize=(10,10))

df.plot(stacked=True, figsize = (10, 5))
plt.xlabel
plt.savefig('level_month')
plt.show()
```
